```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

```

```python
data = pd.read_csv(r"c:\Users\v_bha\Downloads\pca.csv")

print(data)

print(data.columns)
```

```python
print(data.isnull().sum())
```

```python
x = data.drop("class", axis=1)#this make all the clomn into this x and removes the output col

y = data["class"]

fill_missing_values = SimpleImputer(strategy = "median")

x = fill_missing_values.fit_transform(x)

```

```python
scaler = StandardScaler()

x = scaler.fit_transform(x)
```

```python
print(data)
```

```python
print(data.isnull().sum())
```

```python
#pca

pca_model = PCA(n_components = 0.95)


x_pca = pca_model.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size = 0.02,random_state = 100)
```

```python
print(x_train.shape)
print(x_train[:5])
```

```python
model = RandomForestClassifier(n_estimators = 100,criterion = "gini",max_depth = 3,min_samples_leaf = 2,min_samples_split = 5,random_state = 100,bootstrap = True,max_features = "sqrt")


model.fit(x_train,y_train)

model_predictions = model.predict(x_test)
```

```python
#evalution metrices

accuracy = accuracy_score(y_test,model_predictions)
print(accuracy)



cf_matrix = confusion_matrix(y_test,model_predictions)
print(cf_matrix)

classification_reports = classification_report(y_test,model_predictions)
print(classification_reports)
```

```python
cross_score = cross_val_score(model,x,y,cv = 5,scoring = "accuracy")

print(cross_score)

print(cross_score.mean())
```

```python
#hyper paramter tuning

paramter = {"n_estimators":[50,100,200],"criterion":["gini","entropy"],"max_depth" :[5,10,15,200],"min_samples_split": [2,5,10],"min_samples_leaf" : [2,5,10,15],"max_features" : ["sqrt","log2"],"bootstrap":[True,False]}

grid_search = RandomizedSearchCV(RandomForestClassifier(),param_distributions = paramter,cv = 5,scoring = "r2")

grid_search.fit(x_train,y_train)

print(grid_search.best_params_)

print(grid_search.best_score_)

best_model = grid_search.best_estimator_




best_model_predictions = best_model.predict(x_test)

```
