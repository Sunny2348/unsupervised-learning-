```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score

from sklearn.model_selection import ParameterGrid
```

```python
data = pd.read_csv(r"c:\Users\v_bha\Downloads\Wholesale_Data_with_Missing_Values (2).csv")

print(data)
```

```python
print(data.isnull().sum())
```

```python
#hndling missing values

fill_missing_values = SimpleImputer(strategy = "mean")

data = fill_missing_values.fit_transform(data)
```

```python
#now scaling for the columns

scaling = StandardScaler()

data = scaling.fit_transform(data)
```

```python
#first model without dbscan

model = DBSCAN(eps = 5,min_samples = 5)

output = model.fit_predict(data)
```

```python
noise_count = np.sum(output == -1)

print("Noise Points :", noise_count)
```

```python
#unique clusters

unique_clusters  = set(output)

if len(unique_clusters - {-1}) >=2:
    
    silhouette = silhouette_score(data,output)

print(silhouette)
```

```python
#hyperparameter tuning

parameter = {"eps" : [5,6,7,8,9,10],"min_samples" : [3,4,5,6,7,8]}

best_score = -1

best_parameters  = None

best_model = None

for i in ParameterGrid(parameter):

    current_model = DBSCAN(eps = i["eps"],min_samples = i["min_samples"])

    current_model_output = current_model.fit_predict(data)

    score = silhouette_score(data,current_model_output)

    if score > best_score:

        best_score = score

        best_parameters = i

        best_model = current_model


print(best_score)

print(best_parameters)

print(best_model)

```
