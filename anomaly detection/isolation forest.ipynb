```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import IsolationForest

from sklearn.model_selection import ParameterGrid


from sklearn.metrics import accuracy_score,precision_score,f1_score
```

```python
data = pd.read_csv(r"c:\Users\v_bha\Downloads\embedded_system_network_security_dataset.csv")

print(data)
```

```python
#col and out col

x = data.drop(columns = ["label"])

y = data["label"]
```

```python
#hndling missing values

fill_missing_values = SimpleImputer(strategy = "mean")

data = fill_missing_values.fit_transform(data)
```

```python
#now scaling for the columns

scaling = StandardScaler()

x = scaling.fit_transform(x)
```

```python
#model

model = IsolationForest(n_estimators = 100,contamination = 0.05,max_samples = "auto",random_state = 100)
#this contamination=0.05 means we will get anomily score and we sort them we 0.05 measns 1000*0.05 means around 50 col of anokmuly
#max_samopke means for each tree we are assiging max_sample number of trees

output = model.fit_predict(x)

print(output)

#1 means it is normal values 
#if -1 means it is anomaly values
```

```python
pred = []

for x in output:

    if x == -1:
        pred.append(1)

    else:
        pred.append(0)

print(pred)
```

```python
#Model evlaution 

accuracy = accuracy_score(y,pred)

print(accuracy)

precision = precision_score(y,pred)

print(precision)

f1score = f1_score(y,pred)

print(f1_score)
```

```python
X = data[:, :-1]
y = data[:, -1]

print(X.shape)
print(y.shape)


print(type(data))
print(data.shape)

print(type(x))
print(x)
```

```python
#hyperpramter tuning


parameter = {"n_estimators" :[50,100,200],"contamination" : [0.01,0.03,0.05,0.1]}

best_score = -1
best_parameter = None
best_model = None

for i in ParameterGrid(parameter):

    current_model = IsolationForest(n_estimators = i["n_estimators"],contamination = i["contamination"],random_state = 100)

    current_model_output = current_model.fit_predict(X)

    current_model_pred = []
    
    for value in current_model_output:
        
        if value == -1:
            current_model_pred.append(1)
        else:
            current_model_pred.append(0)

    current_model_score = f1_score(y,current_model_pred)

    if current_model_score> best_score:

        best_score = current_model_score

        best_parameter = i

        best_model = current_model


print(best_score)
print(best_parameter)
print(best_model)

    

```

```python
#final best model output 

final_best_model_output = best_model.fit_predict(X)

final_model_prediction = []

for nigga in final_best_model_output:

    if nigga == -1:
        final_model_prediction.append(1)

    else:
        final_model_prediction.append(0)

print(final_model_prediction)
```
